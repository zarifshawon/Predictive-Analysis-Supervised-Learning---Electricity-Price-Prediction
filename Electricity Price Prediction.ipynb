```python
from sklearn.model_selection import train_test_split

df_new = df_new.fillna(df_new.mean())


# extract features and target variable
X1 = df_new[['ForecastWindProduction','ORKTemperature'	,'ORKWindspeed','CO2Intensity',	'ActualWindProduction','SystemLoadEP2','SMPEA', 'HolidayFlag',	'DayOfWeek',	'WeekOfYear',	'Day',		'Year',	'PeriodOfDay']]
y1 = df_new['SMPEP2']



X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor

# Create and train the Decision Tree Classifier
dtc = DecisionTreeRegressor(random_state=42)
dtc.fit(X_train1, y_train1)


```

```python
# Evaluate the performance of the classifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, r2_score, explained_variance_score


y_pred = dtc.predict(X_test1)
mse = mean_squared_error(y_test1, y_pred)
# Evaluate the performance of the Decision Tree Regressor
y_pred_dtc = dtc.predict(X_test1)
mse_dtc = mean_squared_error(y_test1, y_pred_dtc)
r2_dtc = r2_score(y_test1, y_pred_dtc)
evs_dtc = explained_variance_score(y_test1, y_pred_dtc)

print('Decision Tree Regressor Mean Squared Error: ', mse_dtc)
print('Decision Tree Regressor R^2 Score: ', r2_dtc)
print('Decision Tree Regressor Explained Variance Score: ', evs_dtc)

```

```python

from sklearn.svm import SVR
# Create and train the Support Vector Machine
svm = SVR()
svm.fit(X_train1, y_train1)

# Evaluate the performance of the Support Vector Machine
y_pred_svm = svm.predict(X_test1)
mse_svm = mean_squared_error(y_test1, y_pred_svm)
r2_svm = r2_score(y_test1, y_pred_svm)
evs_svm = explained_variance_score(y_test1, y_pred_svm)

print('Support Vector Machine Mean Squared Error: ', mse_svm)
print('Support Vector Machine R^2 Score: ', r2_svm)
print('Support Vector Machine Explained Variance Score: ', evs_svm)

```
